```python
import csv
import pandas as pd
import numpy as np
import pycountry
```

```python
csv_path = "https://raw.githubusercontent.com/Madhuram-S/Thermal_Warming/Guy/NaN_Replaced_Temps.csv"

#read csv
Country_Names = pd.read_csv(csv_path, low_memory=False)
Country_Names.head()
```

```python
input_countries = (['Country'])

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_2

codes = [countries.get(country, countries) for country in input_countries]

#Change Country name to uppercase
codes = dict((k.upper(),c) for k,c in codes[0].items())

print(codes)
```

```python
# Change country names to upper case and trim whitespaces
Country_Names['Country'] = Country_Names['Country'].apply(str.upper)
Country_Names['Country'] = Country_Names['Country'].apply(str.strip)

#map codes with countries
Country_Names['Codes'] = Country_Names['Country'].map(codes)
```

```python
# do unique values exist?
Country_Names[Country_Names.isnull().any(axis = 1)]['Country'].unique()
```

```python
#Function to perform partial matching of country and return code

def returnCountryCode(c_nm):
    #ignore the value in the () for country name
    c_nm = c_nm.split("(",1)[0]
    c_nm = c_nm.strip()
    c_code = [v for k,v in codes.items() if (c_nm.upper() in k.upper()) or (k.upper() in c_nm.upper())]
    if(c_code):
        return c_code[0]
    else:
        return np.nan
    
```

```python
returnCountryCode('REPUBLIC OF KOREA') # test
```

```python
# Apply partial matching to fix the missing codes
Country_Names['Codes'] = Country_Names['Country'].apply(lambda c: returnCountryCode(c))
Country_Names.head()
```
