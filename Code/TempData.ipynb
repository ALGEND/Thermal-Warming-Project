```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import urllib.request as ur

#pd.set_option('display.max_columns', 500)
#csv = requests.get("http://site-ftp.site.com/test/cat/filename.csv")
url="ftp://ftp.ncdc.noaa.gov/pub/data/ghcn/v3/csv/ghcnm.tavg.v3.3.0.20181219.qcu.dat.csv"
req = ur.Request(url) 
csvfile = ur.urlopen(req)

####################
# Warning
####################
#due to file size over 120M, this step will take over 5 minutes to show result.
df = pd.read_csv(csvfile, header=None)
print(df)
```

```python
#drop columns, specify axis=1 for columns and axis=0 for rows
df.drop(df.columns[[4,5,6,8,9,10,12,13,14,16,17,18,20,21,22,24,25,26,28,29,30,
                    32,33,34,36,37,38,40,41,42,44,45,46,48,49,50]], axis=1, inplace=True)
print(df)
```

```python
#rename columns
df.columns = ['ID','year','type','jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
print(df)
```

```python
#checking each column data type
df.info()
```

```python
#conver ID column from int to string type
df['ID'] = df['ID'].apply(str)
df.info()
```

```python
df.to_csv("../data/cleanedtempfile.csv", encoding='utf-8')
```
