```python
#import dependencies
import csv
import pandas as pd
import numpy as np
```

```python
#set path
csv_path = "https://raw.githubusercontent.com/Madhuram-S/Thermal_Warming/Steve_Branch/ghcnm.tavg.v3.3.0.20181219.qcu.dat.csv"

#read csv
avg_temp = pd.read_csv(csv_path, low_memory=False)
avg_temp.head()
```

```python
#convert Year column from str to datetime ** Year
avg_temp['Year'] = pd.to_datetime(avg_temp['Year'], format = '%Y')
#import datetime 
#datetime.datetime.strptime('20-Nov-2002','%d-%b-%Y').strftime('%Y%m%d')
```

```python
#drop all years <1970
avg_temp = avg_temp.drop(avg_temp[avg_temp.Year < '1970'].index)
avg_temp.head()
```
